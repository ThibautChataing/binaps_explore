```python
import pandas as pd
import os
import json
```

```python
#root = r"C:\Users\Thibaut\Documents\These\code\ovh_data"
root = r"C:\Users\Thibaut\Documents\These\code\ovh_data_3"
```

```python
files = os.listdir(root)
patterns = []
logs = []
for f in files:
    if f.endswith('patterns'):
        patterns.append(f)
    else:
        logs.append(f)
print(patterns)
```

```python
first = patterns[0]
print(first)
```

```python
with open(os.path.join(root,first), 'r') as fd:
    s = fd.read()
    j = json.loads("[{}]".format(s.replace('}{', '},{')))
```

```python
df = pd.DataFrame(j)
```

```python
df = df.sort_values(['supp_full', 'supp_half'], ascending=False)
```

```python
df['pat_size'] = df.pat.apply(len)
```

```python
df = pd.DataFrame(columns=['file', 'supp_full', 'supp_half', 'pat'])
dict_file = {}
fid = 0
for file in patterns:
    with open(os.path.join(root,first), 'r') as fd:
        s = fd.read()
        j = json.loads("[{}]".format(s.replace('} {', '},{')))
    df1 = pd.DataFrame(j)
    df1['file'] = fid
    dict_file[fid] = file
    df = pd.concat([df, df1])
    fid += 1
df['pat_size'] = df.pat.apply(len)
```

```python
df = df.sort_values(['supp_full', 'supp_half'], ascending=False)
```
